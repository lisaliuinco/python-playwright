```python
import tkinter as tk
from tkinter import messagebox
import random

class GuessNumberGame:
    def __init__(self):
        self.window = tk.Tk()
        self.window.title("猜數字遊戲")
        self.window.geometry("400x350")
        self.window.resizable(False, False)
        self.window.configure(bg="#2c3e50")

        self.low = 1
        self.high = 100
        self.answer = random.randint(self.low, self.high)
        self.attempts = 0

        self._build_ui()

    def _build_ui(self):
        bg = "#2c3e50"
        fg = "#ecf0f1"
        entry_bg = "#34495e"
        btn_color = "#2980b9"

        tk.Label(self.window, text="猜數字遊戲", font=("Arial", 20, "bold"),
                 bg=bg, fg="#f1c40f").pack(pady=(15, 5))
        tk.Label(self.window, text=f"請猜一個 {self.low} ~ {self.high} 之間的數字",
                 font=("Arial", 11), bg=bg, fg=fg).pack()

        self.attempts_label = tk.Label(self.window, text="已猜次數: 0",
                                       font=("Arial", 10), bg=bg, fg="#95a5a6")
        self.attempts_label.pack()

        self.entry = tk.Entry(self.window, font=("Arial", 14), justify="center",
                              bg=entry_bg, fg=fg, insertbackground=fg, width=10)
        self.entry.pack(pady=10)
        self.entry.bind("<Return>", lambda e: self._check())

        tk.Button(self.window, text="猜！", font=("Arial", 12, "bold"),
                  bg=btn_color, fg="white", activebackground="#3498db",
                  width=12, command=self._check).pack()

        self.hint_label = tk.Label(self.window, text="", font=("Arial", 12),
                                   bg=bg, fg="white")
        self.hint_label.pack(pady=5)

        tk.Button(self.window, text="重新開始", font=("Arial", 10),
                  bg="#27ae60", fg="white", activebackground="#2ecc71",
                  command=self._reset).pack(pady=(5, 10))

    def _check(self):
        guess_str = self.entry.get().strip()
        try:
            guess = int(guess_str)
        except ValueError:
            messagebox.showerror("輸入錯誤", "請輸入有效的整數！")
            return

        if guess < self.low or guess > self.high:
            messagebox.showwarning("超出範圍", f"請輸入 {self.low} ~ {self.high} 之間的數字")
            return

        self.attempts += 1
        self.attempts_label.config(text=f"已猜次數: {self.attempts}")

        if guess < self.answer:
            self.hint_label.config(text="再大一點！ ↑", fg="#3498db")
        elif guess > self.answer:
            self.hint_label.config(text="再小一點！ ↓", fg="#e74c3c")
        else:
            messagebox.showinfo("恭喜", f"答對了！答案是 {self.answer}\n用了 {self.attempts} 次")
            self._reset()

    def _reset(self):
        self.answer = random.randint(self.low, self.high)
        self.attempts = 0
        self.attempts_label.config(text="已猜次數: 0")
        self.hint_label.config(text="", fg="white")
        self.entry.delete(0, tk.END)

    def run(self):
        self.window.mainloop()

GuessNumberGame().run()
```
